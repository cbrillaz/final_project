```python
pbp = []
game_number = 0
# Chris kwy = va6w6hett4j3f2sa76xj5uyx

for game in game_ids:
    play_by_play_url = f'http://api.sportradar.us/nfl/official/trial/v5/en/games/{game}/pbp.json?api_key=va6w6hett4j3f2sa76xj5uyx'
    play_by_play = requests.get(play_by_play_url)
    time.sleep(1.1)
    jsoned = play_by_play.json()
    if jsoned["status"] != "postponed":
        q1 = jsoned["periods"][0]
        q2 = jsoned["periods"][1]
        q3 = jsoned["periods"][2]
        q4 = jsoned["periods"][3]

        game_play_number = 0

        quarters = [q1, q2, q3, q4]
        for quarter in quarters: #Begin Quarter Level
            for drive in range(0, len(quarter["pbp"])): #Begin Drive Level
                drive_play_number = 0
                if quarter["pbp"][drive]["type"] == "drive":
                    for play in range(0,len(quarter["pbp"][drive]["events"])): #Begin Play Level
                        if quarter["pbp"][drive]["events"][play]["type"] == 'play':
                            if (quarter["pbp"][drive]["events"][play]["play_type"] == "rush" or quarter["pbp"][drive]["events"][play]["play_type"] == "pass") and quarter["pbp"][drive]["events"][play]["play_type"] != "field_goal" and quarter["pbp"][drive]["events"][play]["play_type"] != "conversion" and quarter["pbp"][drive]["events"][play]["play_type"] != "kick_off" and quarter["pbp"][drive]["events"][play]["play_type"] != "penalty":
                                if "deleted" not in quarter["pbp"][drive]["events"][play]:
                                    play_dict = {}
                                    play_dict["weather"] = jsoned.get("weather", "")
                                    play_dict["week_number"] = jsoned["summary"]["week"].get("sequence", "")
                                    play_dict["time_zone"] = jsoned.get("utc_offset", "")
                                    play_dict["game_time"] = jsoned.get("scheduled")
                                    play_dict["year"] = jsoned["summary"]["season"].get("year")
                                    play_dict["stadium"] = jsoned["summary"]["venue"].get("name")
                                    play_dict["surface"] = jsoned["summary"]["venue"].get("surface")
                                    play_dict["roof"] = jsoned["summary"]["venue"].get("roof_type")
                                    play_dict["home_team"] = jsoned["summary"]["home"].get("name")
                                    play_dict["away_team"] = jsoned["summary"]["away"].get("name")
                                    game_play_number += 1
                                    play_dict["game_play_number"] = game_play_number
                                    play_dict["quarter"] = quarter.get("sequence")
                                    play_dict["drive_number"] = quarter["pbp"][drive].get("sequence")
                                    play_dict["start_reason"] = quarter["pbp"][drive].get("start_reason") #Identifies why a team started a drive; i.e. fumble, kickoff, punt or interception
                                    play_dict["end_reason"] = quarter["pbp"][drive].get("end_reason") #Identifies why a team ended a drive
                                    play_dict["duration"] = quarter["pbp"][drive].get("duration") # Drive Duration
                                    play_dict["plays_in_drive"] = quarter["pbp"][drive].get("play_count")
                                    play_dict["drive_play_number"] = drive_play_number
                                    play_dict["play_type"] = quarter["pbp"][drive]["events"][play].get("play_type") #Rush or Pass
                                    if "details" in quarter["pbp"][drive]["events"][play]:
                                        play_dict["play_result"] = quarter["pbp"][drive]["events"][play]["details"][0].get("category")
                                    play_dict["game_clock"] = quarter["pbp"][drive]["events"][play].get("clock") #Game Time Remaining
                                    play_dict["play_clock"] = quarter["pbp"][drive]["events"][play].get("play_clock")
                                    play_dict["wall_clock"] = quarter["pbp"][drive]["events"][play].get("wall_clock")
                                    play_dict["home_points"] = quarter["pbp"][drive]["events"][play].get("home_points")
                                    play_dict["away_points"] = quarter["pbp"][drive]["events"][play].get("away_points")
                                    play_dict["down"] = quarter["pbp"][drive]["events"][play]["start_situation"].get("down")
                                    play_dict["distance"] = quarter["pbp"][drive]["events"][play]["start_situation"].get("yfd") #yards for first down
                                    play_dict["offense"] = quarter["pbp"][drive]["events"][play]["start_situation"]["possession"].get("name") #who has possession
                                    play_dict["field_position"] = quarter["pbp"][drive]["events"][play]["start_situation"]["location"].get("name") #whos side of the field they are on
                                    play_dict["line_of_scrimmage"] = quarter["pbp"][drive]["events"][play]["start_situation"]["location"].get("yardline") #what yardline they are on
                                    play_dict["men_in_box"] = quarter["pbp"][drive]["events"][play].get("men_in_box")
                                    play_dict["screen_pass"] = quarter["pbp"][drive]["events"][play].get("screen_pass")
                                    play_dict["blitz"] = quarter["pbp"][drive]["events"][play].get("blitz")
                                    play_dict["play_direction"] = quarter["pbp"][drive]["events"][play].get("play_direction", "")
                                    play_dict["qb_at_snap"] = quarter["pbp"][drive]["events"][play].get("qb_at_snap")
                                    play_dict["huddle"] = quarter["pbp"][drive]["events"][play].get("huddle")
                                    if quarter["pbp"][drive]["events"][play]["play_type"] == 'rush' and quarter["pbp"][drive]["events"][play]["statistics"][0]["stat_type"] != "timeout":
                                        rb_index = next((index for (index, d) in enumerate(quarter["pbp"][drive]["events"][play]["statistics"]) if d["stat_type"] == "rush"), None)
                                        drive_play_number += 1
                                        play_dict["flex_name"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index]["player"].get("name")
                                        play_dict["flex_position"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index]["player"].get("position")
                                        play_dict["flex_team"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index]["team"].get("name")
                                        play_dict["flex_attempt"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("attempt") #the attempt number for the RB
                                        play_dict["flex_yards"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("yards")
                                        play_dict["flex_running_lane"] = quarter["pbp"][drive]["events"][play].get("running_lane", "")
                                        play_dict["flex_firstdown"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("firstdown")
                                        play_dict["flex_inside20"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("inside_20")
                                        play_dict["flex_goaltogo"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("goaltogo")
                                        play_dict["flex_brokentackle"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("broken_tackles")
                                        play_dict["flex_kneel"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("kneel_down")
                                        play_dict["flex_scramble"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("scramble")
                                        play_dict["flex_yacontact"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("yards_after_contact")
                                        play_dict["touchdown"] = quarter["pbp"][drive]["events"][play]["statistics"][rb_index].get("touchdown")
                                    elif quarter["pbp"][drive]["events"][play]["play_type"] == 'pass' and quarter["pbp"][drive]["events"][play]["statistics"][0]["stat_type"] != "timeout":
                                        if "nullified" not in quarter["pbp"][drive]["events"][play]["statistics"][0]:
                                            qb_index = next((index for (index, d) in enumerate(quarter["pbp"][drive]["events"][play]["statistics"]) if d["stat_type"] == "pass"), None)
                                            wr_index = next((index for (index, d) in enumerate(quarter["pbp"][drive]["events"][play]["statistics"]) if d["stat_type"] == "receive"), 0)
                                            db_index = next((index for (index, d) in enumerate(quarter["pbp"][drive]["events"][play]["statistics"]) if (d["stat_type"] == "defense" and d["def_target"] == 1)), 0)
                                            drive_play_number += 1
                                            play_dict["players_rushed"] = quarter["pbp"][drive]["events"][play].get("players_rushed")
                                            play_dict["pocket_location"] = quarter["pbp"][drive]["events"][play].get("pocket_location")
                                            play_dict["pass_route"] = quarter["pbp"][drive]["events"][play].get("pass_route")
                                            play_dict["touchdown"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("touchdown")
                                            play_dict["QB_attempt"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("attempt")
                                            play_dict["QB_complete"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("complete")
                                            play_dict["QB_air_yards"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("att_yards", "") #air yards
                                            play_dict["QB_firstdown"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("firstdown") #first down achieved
                                            play_dict["QB_inside20"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("inside_20")
                                            play_dict["QB_goaltogo"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("goaltogo")
                                            play_dict["QB_blitz"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("blitz")
                                            play_dict["QB_hurry"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("hurry")
                                            play_dict["QB_knockdown"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("knockdown")
                                            play_dict["QB_pocket_time"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("pocket_time")
                                            play_dict["QB_name"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index]["player"].get("name")
                                            play_dict["QB_team"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index]["team"].get("name")
                                            play_dict["QB_id"] = quarter["pbp"][drive]["events"][play].get('id')                                
                                            play_dict["flex_firstdown"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("firstdown", "")
                                            play_dict["flex_target"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("target", "")
                                            play_dict["flex_inside20"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("inside_20", "")
                                            play_dict["flex_goaltogo"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("goaltogo", "")
                                            if "player" in quarter["pbp"][drive]["events"][play]["statistics"][wr_index]:
                                                play_dict["flex_name"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index]["player"].get("name")
                                                play_dict["flex_position"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index]["player"].get("position")
                                            play_dict["flex_team"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index]["team"].get("name")
                                            if "player" in quarter["pbp"][drive]["events"][play]["statistics"][db_index]:
                                                play_dict["DB_name"] = quarter["pbp"][drive]["events"][play]["statistics"][db_index]["player"].get("name")
                                            play_dict["DB_team"] = quarter["pbp"][drive]["events"][play]["statistics"][db_index]["team"].get("name")
                                            play_dict["DB_target"] = quarter["pbp"][drive]["events"][play]["statistics"][db_index].get("def_target")
                                            play_dict["DB_completion"] = quarter["pbp"][drive]["events"][play]["statistics"][db_index].get("def_comp")
                                            play_dict["DB_tackle"] = quarter["pbp"][drive]["events"][play]["statistics"][db_index].get("tackle")

                                            if "complete" not in quarter["pbp"][drive]["events"][play]["statistics"][qb_index]: #INCOMPLETE PASS
                                                play_dict["QB_incomplete_type"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("incompletion_type", "")
                                                play_dict["flex_dropped"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("dropped", "")
                                            elif quarter["pbp"][drive]["events"][play]["statistics"][qb_index]["complete"] == 0: #INCOMPLETE PASS
                                                play_dict["QB_incomplete_type"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("incompletion_type", "")
                                                play_dict["flex_dropped"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("dropped", "")
                                            else: #COMPLETE PASS
                                                play_dict["QB_yards"] = quarter["pbp"][drive]["events"][play]["statistics"][qb_index].get("yards")
                                                play_dict["flex_reception"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("reception")
                                                play_dict["flex_yards"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("yards") #Unique
                                                if "yards_after_catch" in quarter["pbp"][drive]["events"][play]["statistics"][wr_index]:
                                                    play_dict["flex_yacatch"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("yards_after_catch") #Unique
                                                if "yads_after_contact" in quarter["pbp"][drive]["events"][play]["statistics"][wr_index]:
                                                    play_dict["flex_yacontact"] = quarter["pbp"][drive]["events"][play]["statistics"][wr_index].get("yards_after_contact") #Unique
                                    else:
                                        pass

                                    pbp.append(play_dict)
            #                         print(f'Game Plays: {game_play_number}')
        #                             print(f'Quarter: {quarter["sequence"]}')
        #                             print(f'Drive {drive} , Play: {play}')

            #                         print(f'{quarter["pbp"][drive]["events"][play]["id"]}')
            #                         print(f'{quarter["pbp"][drive]["events"][play]["play_type"]}')
        game_number += 1
        print(f'Games Complete: {game_number}')
```
